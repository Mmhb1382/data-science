### Second step: Handling missing values
#### At the first look we'll get that there are a few columns that have a lot of entries with NaN values
#### These columns are: `Workman Joine Data`, `Customer Rating`, `Workman Rating Score`, `Comment`, and `Device`
#### Let's look at the number of missing entries of each of them and then decide what to do.


````python
workman_join_date_null = achare['Workman Joine Date'].isnull().sum()
customer_rating_null = achare['Customer Rating'].isnull().sum()
workman_rating_score_null = achare['workman Rating Score'].isnull().sum()
comment_null = achare['Comment'].isnull().sum()
device_null = achare['Device'].isnull().sum()
print('Number of NaN values in each column:\nColumn: Workman Joine Date: ', workman_join_date_null
      , '\nColumn: Customer Rating: ', customer_rating_null
      , '\nColumn: Workman Rating Score: ', workman_rating_score_null
      , '\nColumn: Comment: ', comment_null
      , '\nColumn: Device: ', device_null)
````

#### We can see that those numbers are huge and in the other hand, the number of dataset rows are: 15353
#### So we can say that these columns must be filled or dropped.
#### The number of first column `Workman Joine Date` null values are the half of its total entries. So for this column we can drop the rows whose this column value is NaN. But for the other three columns, we drop the whole column due to the massiveness of number. Also for the `Device` column there are only 79 empty entries so we just drop the rows corresponding to these entries.

````python
achare = achare.drop(columns=['Customer Rating', 'workman Rating Score', 'Comment'])
````

````python
achare = achare.dropna(subset = ['Workman Joine Date', 'Device'])
````

````python
# Apply one-hot encoding on the 'Quoted Task' column
encoder = pd.get_dummies(achare['Quoted Task'], prefix = 'Quoted Task: ')
# Remove the underscores from the new column names
encoder.columns = encoder.columns.str.replace('_', '', regex=False)
achare = achare.drop('Quoted Task', axis=1)
achare = achare.join(encoder)
````

````python
# Apply one-hot encoding on the 'Device' column
encoder = pd.get_dummies(achare['Device'], prefix = 'Device: ')
# Remove the underscores from the new column names
encoder.columns = encoder.columns.str.replace('_', '', regex=False)
achare = achare.drop('Device', axis=1)
achare = achare.join(encoder)
````
